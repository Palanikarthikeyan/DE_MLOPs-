```python
Zscore 
------- 
 |->standard score
      X - mean of the dataset
Z =   ------------------------ 
           standard deviation
X - data point

Zscore = 0 +1 -1 
  +2 or more - High
```

```python
d = np.array([50,60,70,80,90])
mean = np.mean(d)
std_dev = np.std(d)
zscore = (d - mean) / std_dev
zscore
```

```python
d[np.abs(zscore) > 2]
```
